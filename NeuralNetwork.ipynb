```python
import pandas as pd
```

```python
B_data = pd.read_excel ('/content/Regression.xlsx' )
B_data
```

```python
y = B_data[ 'Earnings' ]
```

```python
x = B_data[ ['Cost', 'Grad', 'Debt', 'City'] ]
```

```python
from keras.models import Sequential as Sq
from keras.layers import Dense, Activation
```

```python
##First defined the structure of the model
ElvisNNModel = Sq()

##After defining the structure, add the input and hidden layers

ElvisNNModel.add( Dense(32, activation='relu' , input_dim = 4 ) )
ElvisNNModel.add( Dense(32, activation='relu') )
ElvisNNModel.add( Dense(1) )
```

```python
ElvisNNModel.compile( loss='mean_squared_error', optimizer='adam', metrics=['accuracy'] ) 
```

```python
ElvisNNModel.fit(x,y,epochs=1000)
```

```python
ElvisNNModel.predict( [[22000,80,70,1]] )
```
